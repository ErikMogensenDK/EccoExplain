```python
def normalize(components):
	range = np.ptp(components)
	min = np.min(components)
	normalized_components = [int(np.round(((x - min)/range)*10)) for x in components]
	normalized_components = np.around(normalized_components)
	return np.around(normalized_components, decimals=0)
```

```python
normalized_components = [normalize(component) for component in explainer.components]
```

```python
def create_new_result_string(normalized_components):
	new_strings = []
	for component in normalized_components:
		new_string = 'factor: \nActivations\n<start>\n'

		for i in range(len(component)):
			new_string = new_string + f'{explainer.tokens[i]}\t{component[i]}\n'
		new_string = new_string + '<end>\n'

		new_string = new_string + 'same_string but with all zeros filtered out:\n<start>'	
		for i in range(len(component)):
			if component[i]>0:
				new_string = new_string + f'{explainer.tokens[i]}\t{component[i]}\n'

		new_string = new_string +'<end>\n'
		new_strings.append(new_string)
	return new_strings
```
